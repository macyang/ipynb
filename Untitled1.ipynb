```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
from IPython.html.widgets import interact
```

```python
from mall.Vertex import *
from mall.Graph import *
from mall.Map import *
from mall.Dot import *

map = Map()
```

```python
def plot_dots(T):
    xpos = []
    ypos = []
    dL = []
    for dId in range(500):
        dL.append(Dot(map, 2, dId))
    for d in dL:
        for t in range(T):
            d.walk()
        xpos.append(d.getX())
        ypos.append(d.getY())
    plt.xlim(0.0, 6.0)
    plt.ylim(0.0, 5.0)
    plt.scatter(xpos, ypos)
    plt.show()
```

```python
interact(plot_dots, T=(0,10));
```
