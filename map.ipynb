### Visualize device location data on Google map.

```python
%%javascript
require.config({
    paths: {async: '//cdnjs.cloudflare.com/ajax/libs/requirejs-async/0.1.1/async'}
});
```

```python
# phonenum = 'ph3a8305f4fdc11289c8a0c2039c09e366c1269364'
df = pd.read_csv('loc_samples.csv')
```

```python
def display_loc_markers_map(df, div):    
    loc_markers_template = jinja2.Template(
    """
    require(['async!http://maps.googleapis.com/maps/api/js?sensor=false&libraries=visualization&callback=console.log'], function() {
        var mapOptions = {
            zoom: 10.0,
            center: new google.maps.LatLng(37.575, -122.4183333),
            mapTypeId: google.maps.MapTypeId.ROADMAP
        };
        var map = new google.maps.Map(document.getElementById('loc_markers'), mapOptions);
        var markers = Array();

        {% for d in loc_data %}
        markers.push(new google.maps.Marker({
            position: new google.maps.LatLng({{ d[1] }}, {{ d[2] }}),
            map: map
        }));
        {% endfor %}
    });    
    """
    )
    display(Javascript(loc_markers_template.render(loc_data=df.itertuples())))
```

```python
i = interact(
    display_loc_markers_map,
    df=widgets.fixed(df),
    div=widgets.HTMLWidget(value='<div id="loc_markers" style="width: 800px; height: 600px"></div>')
)
```
