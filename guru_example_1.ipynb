# Example 1
### In this example, we will generate large number of random data points and then calculate the basic stats of the generated dataset.

## Create an Application Context
An application context is basically a SparkConext wrapper. It is referenced by name and is required for all GURU API calls.
Here we create a new context named "guru-context". 

```python
resp = createContext("guru-context")
print resp
```

## Generate some random data for testing
In this example, we invoke the <b>RandomData</b> module to generate 1,000,000 data points. The random data generated by the module is made available in a RDD named "somedata" (see the <b>output0</b> config parameter). Later on, we will reference this RDD by its name and use it as input to other modules.

```python
conf = """
RandomData {
    mode = "normalVector"
    numRows = 1000000
    numCols = 2
    output0 = "somedata"
}
"""
resp = run("guru-context", "tap.engine.RandomData", conf)
print resp
```

## Check the stats of the generated dataset
Here we invoke the <b>SummaryStats</b> module to gather basic stats of the generated dataset. Note that it uses the RDD generated by the <b>RandomData</b> module as its input (see the <b>input0</b> config parameter).

```python
conf = """
SummaryStats {
    input0 = "somedata"
}
"""
resp = run("guru-context", "tap.engine.SummaryStats", conf)
print resp
```

## Clean up
Delete the application context to free up the resources associated with it.

```python
resp = deleteContext("guru-context")
```
