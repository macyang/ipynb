# Example 1
### In this example, we will generate 1,000,000 random data points, calculate the basic stats of the generated dataset, and also visualize the dataset using sampled data.

## Create an Application Context
An application context is basically a SparkConext wrapper. It is referenced by name and is required for all GURU API calls.
Here we create a new context named "guru-context" using the <b>/contexts</b> REST end point. 

```python
q = contextAPI + '/' + 'guru-context?num-cpu-cores=4&mem-per-node=512m'
requests.post(q)
```

## Generate some random data for testing
In this example, we invoke the <b>RandomData</b> module to generate 1,000,000 data points. In this example, the random data generated by the module is made available in a RDD named "somedata" (see the <b>output0</b> config parameter).

```python
conf = """
RandomData {
    output0 = "somedata"
    mode = "normalVector"
    numRows = 1000000
    numCols = 2
}
"""
job = jobAPI + '?appName=tap&classPath=tap.engine.RandomData&context=guru-context&sync=true'
r = requests.post(job, data = conf)
resp = json.loads(r.text)
print resp['result']
```

## Check the stats of the generated dataset
Here we invoke the <b>SummaryStats</b> module to gather basic stats of the generated dataset.

```python
conf = """
SummaryStats {
    input0 = "somedata"
}
"""
job = jobAPI + '?appName=tap&classPath=tap.engine.SummaryStats&context=guru-context&sync=true'
r = requests.post(job, data = conf)
resp = json.loads(r.text)
print resp['result']
```

## Visualization using random samples from the dataset
Here we invoke the <b>TakeSample</b> module to fetch 100 random samples from the dataset, and render them in a scatter plot.

```python
%matplotlib inline
```

```python
conf = """
TakeSample {
    input0 = "somedata"
    count = 100
    seed = 123456
}
"""
job = jobAPI + '?appName=tap&classPath=tap.engine.TakeSample&context=guru-context&sync=true'
r = requests.post(job, data = conf)
resp = json.loads(r.text)
data_list = resp['result']['data']
# format the data for matplot
xlist = []
ylist = []
for dS in data_list:
    d = json.loads(dS)
    xlist.append(d[0])
    ylist.append(d[1])
plt.scatter(xlist, ylist, marker='o', color='r');
```

## Clean up
Delete the application context to free up the resources associated with it.

```python
q = contextAPI + '/' + 'guru-context'
requests.delete(q)
```
